# Python urllib2模块获取http请求

## 1. 导入urllib2模块

```python
#-*- coding: UTF-8 -*-
# urllib2在Python3中被替换为 urllib.request。
import urllib2
```

## 2. 使用urllib2获取页面所有信息

```python
url = "http://www.8kraw.com/"
print urllib2.urlopen(url).read() #读取网页全部信息
```

# 3. 使用urllib2获取页面所有信息，并返回由每一行组成的列表

```python
url = "http://www.8kraw.com"
print urllib2.urlopen(url).readlines() #以列表形式返回网页全部信息
```

# 4. 逐行获取页面信息

```python
url = "http://www.8kraw.com/"
response = urllib2.urlopen(url)
while True:
    #  读取每一行的网页数据，并压入列表
    line = response.readline()
    if not line:
        break
    print line
```
